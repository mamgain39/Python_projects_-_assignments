## 2. Check for missing values in all columns and replace them with the appropriate metric
## (Mean/Median/Mode)

```python
ecom.isnull().sum()        # Checking for null values in all columns. We can see that only Description and Quantity have nulls.
```

```python
ecom['Description'].mode()[0]                # Calculating mode for Description column.
```

```python
ecom['Description']=ecom['Description'].fillna(ecom['Description'].mode()[0])   # Replacing nulls in Description with mode.
```

```python
ecom['CustomerID'].mode()[0]                 # Calculating mode for CustomerID column.
```

```python
ecom['CustomerID']=ecom['CustomerID'].fillna(ecom['CustomerID'].mode()[0])      # Replacing nulls in CustomerID with mode.
```

```python
ecom.isnull().sum()                      # Now we can see that there are no null values.
```
